```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv('covid_19_indonesia_time_series_all.csv')
```

```python
data
```

```python
data.head(5)
```

```python
data.info()
```

```python
df = data[data['Location'] != 'Indonesia']
df
```

```python
dfBengkulu = df[df['Location'] == 'Bengkulu']
dfBengkulu
```

```python
data2021 = dfBengkulu[pd.DatetimeIndex(dfBengkulu['Date']).year == 2021]
data2021
```

```python
dataJuli2021 = data2021[pd.DatetimeIndex(data2021['Date']).month == 6]
dataJuli2021
```

```python
from sklearn.linear_model import LinearRegression
```

```python
tanggal = pd.DatetimeIndex(dataJuli2021['Date']).day.values
newCase = dataJuli2021['New Cases'].values

x_train = tanggal[:,np.newaxis]
y_train = newCase

lm = LinearRegression()
lm.fit(x_train, y_train)
```

```python
pb  = lm.predict(x_train)
dfc = pd.DataFrame({'x': tanggal,'y': pb})

plt.scatter(tanggal, newCase)
plt.plot(dfc['x'], dfc['y'], color='red', linewidth=2)
plt.xlabel('Tanggal')
plt.ylabel('Jumlah Kasus')
plt.show()
```
